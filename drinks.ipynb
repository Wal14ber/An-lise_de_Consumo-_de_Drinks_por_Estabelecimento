```python
#importar as bibliotecas necessárias ex. pandas, numpy, matplotlib
import pandas as pd
```

```python
#ler planilha de clientes atendidos por Canal 1ª tabela
df1 = pd.read_excel('dados.xlsx', sheet_name='Por Canal')
df1

```

```python
#ler planilha com quantidade de pedidos
df2 = pd.read_excel('dados.xlsx', sheet_name='Consumo de Produtos')
df2



```

```python
#ler as abas da planilha
import pandas as pd
xls = pd.ExcelFile('dados.xlsx')
print(xls.sheet_names)  # Lista os nomes das abas na planilha
```

```python
#filtrar categorias desejadas
df2 = df2[df2['T. Categoria'].isin(['DRINKS', 'SOFT DRINKS'])]
df2
```

```python
#calcular o total de clientes atendidos por loja
n_pessoas = df1.groupby('Estabelecimento')['N.Pessoas'].sum()
n_pessoas

```

```python
#calcular quantidades de produtos vendidos por loja e categoria
quant = df2.groupby(['Estabelecimento','Categoria'])['Quant.'].sum()
quant
```

```python
#calcular quantidade de pedidos por cliente
quant_por_pessoa = quant / n_pessoas
quant_por_pessoa
```

```python
#plotar grafico para visualizar os resultados
import matplotlib.pyplot as plt
quant_por_pessoa.plot.barh()
quant_por_pessoa.plot.barh()




```

```python
#plotar grafico para visualizar os resultados
ax = quant_por_pessoa.plot.barh()
ax.bar_label(ax.containers[0],fmt='{:.2f}');
#ax.set_xlim(0,1)
ax.set_xlim(0, ax.get_xlim()[1]*1.1);  # Ajusta o limite superior do eixo x para dar espaço às etiquetas  






```

```python
#mostrar top 5 produtos por loja e categoria
top5 = df2.groupby(['Estabelecimento','T. Categoria','Descrição'])['Quant.'].sum()
top5 = top5.sort_values(ascending=False)
top5 = top5.groupby(['Estabelecimento','T. Categoria']).head(5)
top5 = top5.sort_index()
top5





```
